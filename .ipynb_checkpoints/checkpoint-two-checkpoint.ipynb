# About This Dataset

Pokemon Stats including new Scarlet and Violet versions and Updated Stats

**Column Info**

* `'#'`:  Pokedex number (each pokemon has an individual index number registered into the "national pokedex").

* `Name`: Pokemon name

* `Variation`: Pokemon Variation, such as Mega evolution, Regional Form, This can influence in the pokemon's stats.

* `Type1`: Pokemon type 1

* `Type2`: Pokemon Secondary type

* `Total`: Sum of Attack, Sp. Atk, Defense, Sp. Def, Speed and HP

* `HP-Speed`: Stat for each Pokemon/Variation

```python
#Seconday pokemon types used?
poke_df['Type2'].value_counts()
```

```python
poke_df['Type2'].value_counts().count()
```

```python
#creating variables for each Type count for later comparison.
type1_counts = poke_df["Type1"].value_counts()
type2_counts = poke_df["Type2"].value_counts()

type_counts1 = pd.concat([type1_counts], axis = 1)
type_counts1
```

```python
type_counts2 = pd.concat([type2_counts], axis = 1)
type_counts2
```

```python
# Peeking those stats 
poke_df.describe()
```

```python
#What pokemon has the Max Stats Total?
max_total = poke_df.groupby(["Name"]).sum()
max_total.sort_values(by="Total", ascending = False)
```

```python
#What pokemon has the Max Attack stat?
max_attack = poke_df.groupby(["#"]).sum()
max_attack.sort_values(by="Attack", ascending = False)
```
